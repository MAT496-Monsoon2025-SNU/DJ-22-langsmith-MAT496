```python
from dotenv import load_dotenv
load_dotenv(dotenv_path="../../.env", override=True)
```

```python
from langsmith import Client

example_inputs = [
("How many squares are there on a standard chess board?", "There are 64 squares on a standard chess board."),
("Who is the highest rated chess player to ever live?", "The highest rated chess player to ever live is Magnus Carlsen."),
("What is the most famous position in chess history?", "The most famous chess position is often considered the Saavedra position. White wins by underpromoting 6.c8=R!!"),
]

client = Client()
dataset_name = "Chess Q/A"

dataset = client.create_dataset(
  dataset_name=dataset_name, description="History and factual questions regarding chess",
)

inputs = [{"question": input_prompt} for input_prompt, _ in example_inputs]
outputs = [{"output": output_answer} for _, output_answer in example_inputs]

client.create_examples(
  inputs=inputs,
  outputs=outputs,
  dataset_id=dataset.id,
)
```
